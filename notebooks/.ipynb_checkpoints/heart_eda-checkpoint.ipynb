## Heart Disease EDA
In this notebook we explore [Heart disease data from Kaggle](https://www.kaggle.com/datasets/fedesoriano/heart-failure-prediction/data).
The data has the following variables, taken verbatim from the page:
- Age: age of the patient [years]
- Sex: sex of the patient [M: Male, F: Female]
- ChestPainType: chest pain type [TA: Typical Angina, ATA: Atypical Angina, NAP: Non-Anginal Pain, ASY: Asymptomatic]
- RestingBP: resting blood pressure [mm Hg]
- Cholesterol: serum cholesterol [mm/dl]
- FastingBS: fasting blood sugar [1: if FastingBS > 120 mg/dl, 0: otherwise]
- RestingECG: resting electrocardiogram results [Normal: Normal, ST: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV), LVH: showing probable or definite left ventricular hypertrophy by Estes' criteria]
- MaxHR: maximum heart rate achieved [Numeric value between 60 and 202]
- ExerciseAngina: exercise-induced angina [Y: Yes, N: No]
- Oldpeak: oldpeak = ST [Numeric value measured in depression]
- ST_Slope: the slope of the peak exercise ST segment [Up: upsloping, Flat: flat, Down: downsloping]
- HeartDisease: output class [1: heart disease, 0: Normal]

The goal is to determine the features that influence the target variable (`HeartDisease`) and build a model to predict the probability of a person having heart disease. In this notebook we will do data clean up, explore the variables and their relationships with each other and the target variable. In the next notebook we shall commence the model building. We use a few utility functions which are found in `utils/data_utils.py` for plots and other useful summaries.

### Splitting data
Before proceeding, we split the data into training and testing, using an 80-20 split and stratifying using the target variable to ensure the same proportion in the training and test sets.

```python
all_vars = list(data.columns)
preds = [var for var in all_vars if var != 'HeartDisease']
features = data[preds]
target = data['HeartDisease']
X_train, X_test, y_train, y_test = train_test_split(features, target,test_size=0.2,random_state=1996,stratify=target)
X_train.to_csv('../data/training_features.csv', index = False)
X_test.to_csv('../data/testing_features.csv', index = False)
y_train.to_csv('../data/training_labels.csv', index = False)
y_test.to_csv('../data/testing_labels.csv', index = False)
```

```python
num_vars = ['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']
cat_vars = [var for var in list(X_train.columns) if var not in num_vars]
full_train = pd.concat([X_train, y_train], axis=1)

print(num_vars)
print(cat_vars)
```

### Investigating Cholesterol Issue
We now investigate the cholesterol variable. Serum cholesterol, according to [this page](https://www.healthline.com/health/serum-cholesterol), is a combination of low-densisity cholesterol, high density cholesterol and triglycerides. The ideal situation is low levels of LDL and triglycerides, and higher levels of HDL. Serum cholestorol is caluclated as LDL + HDL + 20% triglycerides. In general, the lower the better for health. However, a value of 0 is unlikely.  We compute the correlations of the continuous variables with the target variable and surprisingly, `Cholesterol` is negatively correlated with heart disease (-0.23). This may be misleading and due to the zeros.  

Below we remove the zero cholestoerol values, and remain with 589 (there are 145 zero cholesterol values). Recalulating the correlations we see that having removed these zeros, cholestorol is now positively correlated with heart disease (which is more in line with our general knowledge). We could remove these zeros from the data on grounds on innacurate data collection, or we may impute using the mean or median. We shall do both and investigate how model performance is affected in the next notebook.

One other thing of note is that 87% of the observations that have a zero cholesterol value have heart disease. This might be misleading when it comes to model building since the algorithms may use this variable to say 'if cholesterol = 0 then predict heart disease', yet we know this value is likely erroneous. So it would be unwise to use these values as they are.

```python
correlations_data_2  = numerical_and_target.corr()
print("Correlation between continuous variables and target: \n")
print(correlations_data_2['HeartDisease'].sort_values())
non_zero_chol = full_train[full_train['Cholesterol']>0].copy()
zero_chol = full_train[full_train['Cholesterol']==0].copy()
num_target = num_vars + ['HeartDisease']
correlations_data_3  = non_zero_chol[num_target].corr()
print("\nCorrelation between continuous variables and target after removing zero cholesterol values: \n")
print(correlations_data_3['HeartDisease'].sort_values())
```

## Loess Diagnostic
We can check for non-linearities between continuous predictors and the binary target using a LOESS diagnostic. This helps us decide whether to include polynomial terms (e.g., `age^2`) when we fit models in the next notebook. Briefly, for a continuous variable `x` and a specific value `x_0`, we perform a locally weighted least squares regression within a defined window around `x_0`. This provides an estimate of `P(Heart Disease | x_0)` based on the local trend in the data. If the resulting smoothed curve shows non-linear patterns, it suggests that a simple linear term may be insufficient—so in the logistic regression that follows, we should consider adding polynomial features.


We perform this loess check for each of the continuous variables below and note quite a few non-linearities in the plots. We shall therefore experiment with polynomial features in feature engineering in the next notebook.

```python
data = X_train.copy()
data['HeartDisease'] = y_train
```

```python
loess_diagnostic_grid( data, num_vars,
    'HeartDisease',
    frac=0.25,
    logistic_C=1.0,
    n_cols=2,
    scatter_alpha=0.15,
    scatter_size=12,
    figsize_per_plot=(6, 4),
)
```
