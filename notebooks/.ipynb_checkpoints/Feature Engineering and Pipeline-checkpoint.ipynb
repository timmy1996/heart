EDA from before indicated that the categorical variables did not heavily influence the price. So perhaphs the missing values here will not hamper model performance too much. Below we compare the mean price per catgopry, including the 'missing' as a category for each varioable. And the results indicate as much. There does not seem to be a major differnce in the means. Going forward we will impute the missing values with the mode just so the algorithms we use can handle the data properly. 

```python


df = pd.concat([categorical_vars,train_labels],axis=1)
cat_vars = list(categorical_vars.columns)
target = 'Trip_Price'  

for var in cat_vars:
    # Fill missing values explicitly
    df_plot = df.copy()
    df_plot[var] = df_plot[var].fillna('Missing')

    # Group to get both mean and count
    agg_df = df_plot.groupby(var)[target].agg(['mean', 'count']).sort_values('mean')

    # Plot
    plt.figure(figsize=(10, 5))
    ax = sns.barplot(x=agg_df.index, y=agg_df['mean'])

    # Annotate each bar with the count
    for i, (cat, row) in enumerate(agg_df.iterrows()):
        ax.text(i, row['mean'], f'n={int(row["count"])}', 
                ha='center', va='bottom', fontsize=9, color='black')

    plt.title(f'Mean {target} by {var} (with counts)')
    plt.ylabel(f'Mean {target}')
    plt.xlabel(var)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
cat_imputer = SimpleImputer(strategy='most_frequent')
encoder = OneHotEncoder()
cat_pipeline = Pipeline([
    ('imputer', cat_imputer),
    ('encoder', encoder)
])
num_pipeline = Pipeline([
('imputer', SimpleImputer(strategy="median")),
('std_scaler', StandardScaler() )])

num_vars = list(numerical_vars.columns)
full_pipeline = ColumnTransformer([
("num", num_pipeline, num_vars),
("cat", cat_pipeline, cat_vars),
])
```

```python
features_transformed = full_pipeline.fit_transform(train_features)
```

```python
lin_reg = LinearRegression()
#lin_reg.fit(features_transformed, train_labels)
scores = cross_val_score(lin_reg, features_transformed, train_labels,
scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
rmse_scores.mean()
```

```python
percentage_error = (rmse_scores / train_labels.mean().item()) * 100
percentage_error
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
import pandas as pd

class EstimatedCostTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, input_columns=None):
        # Must include: Trip_Distance_km, Per_Km_Rate, Trip_Duration_Minutes, Per_Minute_Rate
        self.input_columns = input_columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # Reconstruct a DataFrame if input is a NumPy array
        if not isinstance(X, pd.DataFrame):
            if self.input_columns is None:
                raise ValueError("Column names must be provided when input is a NumPy array.")
            X = pd.DataFrame(X, columns=self.input_columns)

        X = X.copy()

        try:
            X['cost_distance'] = X['Trip_Distance_km'] * X['Per_Km_Rate']
            X['cost_time'] = X['Trip_Duration_Minutes'] * X['Per_Minute_Rate']
            X['est_cost'] = X['cost_distance'] + X['cost_time']
        except KeyError as e:
            raise KeyError(f"Missing one or more required columns: {e}")

        return X
```

```python
num_pipeline_2 = Pipeline([
('imputer', SimpleImputer(strategy="median")),
('cost_features', EstimatedCostTransformer(input_columns=num_vars)),
('std_scaler', StandardScaler() )])
```
