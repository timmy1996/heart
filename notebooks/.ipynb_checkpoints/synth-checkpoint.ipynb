```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import ast
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import cross_val_score
sys.path.append(os.path.abspath(".."))
from utils.data_utils import missing_value_summary
```

```python
data = pd.read_csv('../data/Synthetic_Churn_Dataset.csv')
```

```python
data.shape
```

```python
data.drop(columns='CustomerID', inplace=True)
data.head(2)
```

```python
all_vars = list(data.columns)
preds = [var for var in all_vars if var != 'Churn']
features = data[preds]
target = data['Churn']
X_train, X_test, y_train, y_test = train_test_split(features, target,test_size=0.2,stratify=target)
X_train.to_csv('../data/training_features.csv', index = False)
X_test.to_csv('../data/testing_features.csv', index = False)
y_train.to_csv('../data/training_labels.csv', index = False)
y_test.to_csv('../data/testing_labels.csv', index = False)
```

```python
data['Churn'] = data['Churn'].astype('category')
data['SeniorCitizen'] = data['SeniorCitizen'].astype('category')
for var in numerical_vars:
    data[var] = pd.to_numeric(data[var], errors='coerce')
```

```python
fig, axes = plt.subplots(n, n_cols, figsize=(12, 4 * n))
axes = axes.flatten()

for i, num_var in enumerate(numerical_vars):
    sns.set_style("white")
    ax = axes[i]
    sns.boxplot(data=data, x=num_var, y='Churn',width=0.2,ax=ax)
    #ax.set_title(f'{num_var.replace('_', ' ')} by Churn')
    #ax.set_ylabel('Density')
    ax.grid(True)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
n = len(categorical_vars)
n_cols = 2
n_rows = (n + 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 4 * n_rows))
axes = axes.flatten()

for i, cat_var in enumerate(categorical_vars):
    ax = axes[i]

    # Safe groupby and proportion calculation
    counts = (
        data
        .groupby([cat_var, 'Churn'], observed=True)
        .size()
        .to_frame(name='Count')
        .reset_index()
    )

    counts['Proportion'] = counts.groupby(cat_var)['Count'].transform(lambda x: x / x.sum())

    # Plot
    sns.barplot(data=counts, x=cat_var, y='Proportion', hue='Churn', ax=ax)
    ax.set_title(f'Proportion of Churn by {cat_var.replace("_", " ")}')
    ax.set_xlabel(cat_var.replace('_', ' '))
    ax.set_ylabel('Proportion')
    ax.tick_params(axis='x', rotation=20)
    ax.grid(False)
    sns.despine(ax=ax)

# Remove unused axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.subplots_adjust(hspace=0.4)  # ADD THIS
plt.show()
```

```python
data['InternetService'] = data['InternetService'].fillna("None")
```

```python
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_vars),
    ('cat', OneHotEncoder(), categorical_vars)
])
```

```python
features = [*numerical_vars, *categorical_vars]
features
```

```python
X_train = data[features]
X_train.head()
```

```python
X_train.info()
```

```python
features_transformed = preprocessor.fit_transform(X_train)
```

```python
logistic_reg_model_1 = LogisticRegression()
logistic_reg_model_1.fit(features_transformed,y_train)
```

```python
y_proba = logistic_reg_model_1.predict_proba(features_transformed)
```

```python
pred = pd.DataFrame(logistic_reg_model_1.predict(features_transformed), columns = ['Pred'])
```

```python
preds = pd.DataFrame(y_proba, columns=['Class_0', 'Class_1'])
preds['PredictedClass'] = (preds['Class_1'] > 0.5).astype(int)
preds['PredClass'] = pred
preds['TrueLabel'] = y_train
preds.head(10)
```

```python
scores = cross_val_score(logistic_reg_model_1, features_transformed, y_train, cv=5, scoring='roc_auc')
print("Mean AUC: ", scores.mean())
```
