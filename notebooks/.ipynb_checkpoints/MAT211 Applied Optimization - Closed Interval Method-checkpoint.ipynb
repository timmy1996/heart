# Cross‑Disciplinary Optimization on Closed Intervals
This notebook contains five short, applied examples—each designed for the **Closed Interval Method** workflow:

1. Biology (bacterial growth) — **maximization**
2. Chemistry (Michaelis–Menten with competitive inhibition) — **maximization**
3. Physics (optics: intensity vs. distance) — **maximization**
4. Statistics (MLE for Binomial) — **maximization**
5. Economics (profit from demand & cost) — **maximization**

For each: define the model, plot on its closed interval, do a quick analytic sketch, and compare with a numerical optimizer (`scipy.optimize.minimize_scalar`).



## Closed Interval Method (recap)
Given a continuous function $f$ on $[a,b]$:
1. Find interior **critical points** where $f'(x)=0$ or undefined.
2. Evaluate $f$ at all **candidates**: $a$, $b$, and interior critical points.
3. The largest value is the absolute maximum; the smallest is the absolute minimum.


The code below imports **NumPy**, **Matplotlib**, and **SciPy** — tools for numerical computation, plotting, and optimization.

The helper function `report_max(f, a, b, name="f")` is a wrapper that:

- Uses `scipy.optimize.minimize_scalar` to **maximize** a function `f` on the interval `[a,b]`.  
  (Since `minimize_scalar` only *minimizes*, we pass `-f(x)` to turn maximization into minimization.)
- Extracts the optimal input `x_star` (where the maximum occurs) and the corresponding maximum value `f_star`.
- Prints a formatted report showing the results, plus some info from the optimizer (success flag and number of evaluations).
- Returns these values so they can be used later in the notebook.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
#these are just some helper functions 
def report_max(f, a, b, name="f"):
    res = minimize_scalar(lambda x: -f(x), bounds=(a,b), method="bounded")
    x_star = float(res.x)
    f_star = float(f(x_star))
    print(f"[Maximize] {name}: x* approx {x_star:.6f}, {name}(x*) approx {f_star:.6f}  | success={res.success}, evals={res.nfev}")
    return x_star, f_star, res
```

If we had a minimization problem, we would use `report_min` as defined below.

```python
def report_min(f, a, b, name="f"):
    res = minimize_scalar(lambda x: f(x), bounds=(a,b), method="bounded")
    x_star = float(res.x)
    f_star = float(f(x_star))
    print(f"[Minimize] {name}: x* approx {x_star:.6f}, {name}(x*) approx {f_star:.6f}  | success={res.success}, evals={res.nfev}")
    return x_star, f_star, res
```

The helper function `plot_on_interval(f, a, b, label, x_star=None, y_star=None, title="")` makes a quick plot of a function on a chosen interval:

- It uses `np.linspace(a, b, 100_001)` to create a fine grid of 100,001 points between `a` and `b`.  
- It evaluates the function `f` on that grid and plots the curve.  
- If an optimal point `(x_star, y_star)` is provided, it marks it with a dot and a dashed vertical line.  
- It labels the axes, adds a title and legend, and shows a grid for readability.  

This helper makes it easy to **visualize functions on their domain** and highlight where the maximum (or minimum) occurs.


```python
def plot_on_interval(f, a, b, label, x_star=None, y_star=None, title=""):
    X = np.linspace(a, b, 100_001)
    Y = f(X)
    plt.figure()
    plt.plot(X, Y, label=label)
    if x_star is not None and y_star is not None:
        plt.scatter([x_star], [y_star], zorder=3, label="Optimizer result")
        plt.axvline(x_star, linestyle="--")
    plt.xlim(a, b)
    plt.xlabel("x")
    plt.ylabel(label)
    plt.title(title if title else label)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
```


## 1) Biology — Bacterial growth (maximize population size)

Model:  
$$
P(t) = 100\,t^2 e^{-0.2 t}, \quad 0 \le t \le 24.
$$
Before we do the python implemenation, below we do the calculus approach as discussed in the lecture.
Find $P'(t)$:
$$
P'(t)
= 100\Big(2t\cdot e^{-0.2t} + t^2 \cdot (-0.2)e^{-0.2t}\Big) 
= 100\,e^{-0.2t}\big(2t - 0.2t^2\big) 
= 20\,e^{-0.2t}\,t(10 - t).
$$

Set $P'(t)=0$:
$$
20\,e^{-0.2t}\,t(10-t)=0 \;\Longrightarrow\; t=0 \;\text{ or }\; t=10.
$$

Thus, on the closed interval $[0,24]$, the **candidates** are
$$
t=0,\quad t=10,\quad t=24 \quad \text{(endpoints + interior critical point).}
$$

**Evaluate \(P\) at all candidates**
Left endpoint:
$$
P(0)=100\cdot 0^2 \cdot e^{0}=0.
$$

Interior critical point:
$$
P(10)=100\cdot 10^2 \cdot e^{-0.2\cdot 10}
=10000\,e^{-2}\approx 1353.3528.
$$

Right endpoint:
$$
P(24)=100\cdot 24^2 \cdot e^{-0.2\cdot 24}
=57600\,e^{-4.8}\approx 474.0334.
$$


Then we Compare the values:
$$
P(0)=0,\qquad P(10)=10000e^{-2}\ (\approx 1353.35),\qquad P(24)=57600e^{-4.8}\ (\approx 474.03).
$$

Therefore, the **absolute maximum** on \([0,24]\) occurs at
$$
\boxed{t^* = 10\ \text{hours},\qquad P_{\max} = 10000\,e^{-2} \approx 1353.35\ }
$$


Candidates: $t\in\{0,10,24\}$. The interior critical point $t=10$ is the maximizer.

**Compare these results with the results from the python implemenation below:**

```python

def P_bact(t):
    return 100.0*(t**2)*np.exp(-0.2*t)

a2, b2 = 0.0, 24.0
t2_star, P_max, res2 = report_max(P_bact, a2, b2, name="P")
plot_on_interval(P_bact, a2, b2, label="P(t)", x_star=t2_star, y_star=P_max,
                 title="Biology: Bacterial population (maximize on [0,24])")

cands_t = np.array([a2, 10.0, b2])
cands_P = P_bact(cands_t)
j = int(np.argmax(cands_P))
print(f"Closed-interval candidates -> t = {cands_t}, P(t) = {cands_P}")
print(f"Max at t = {cands_t[j]:.6f}, P = {cands_P[j]:.6f}")
```


## 2) Chemistry — Michaelis–Menten with competitive inhibition (maximize rate)

Model:  
$$
v(S) = \frac{100S}{(S+2)(1+S/20)}, \quad 0 \le S \le 50.
$$

**Find $v'(S)$:**

Simplify first:
$$
v(S)=\frac{2000S}{(S+2)(S+20)}.
$$

Differentiate:
$$
v'(S)=\frac{2000(40-S^2)}{(S^2+22S+40)^2}.
$$

Set $v'(S)=0$:
$$
40-S^2=0 \;\;\Longrightarrow\;\; S=\sqrt{40}=2\sqrt{10}\ (\approx 6.32).
$$

Thus, on the closed interval $[0,50]$, the **candidates** are
$$
S=0,\quad S=2\sqrt{10},\quad S=50.
$$

**Evaluate $v$ at all candidates**

$$
v(0)=0, \qquad v(2\sqrt{10}) \approx 57.72, \qquad v(50) \approx 27.47.
$$

**Compare the values:**
$$
v(0)=0,\qquad v(2\sqrt{10})\approx 57.72,\qquad v(50)\approx 27.47.
$$

Therefore, the **absolute maximum** on $[0,50]$ occurs at
$$
\boxed{S^* = 2\sqrt{10}\ (\approx 6.32), \qquad v_{\max} \approx 57.72.}
$$

Candidates: $S\in\{0,2\sqrt{10},50\}$. The interior critical point $S=2\sqrt{10}$ is the maximizer.

**Compare these results with the results from the python implemenation below:**

```python

def v_chem(S):
    return 100.0*S / ((S+2.0)*(1.0 + S/20.0))

a, b = 0.0, 50.0
S_star, v_max, res_chem = report_max(v_chem, a, b, name="v")
plot_on_interval(v_chem, a, b, label="v(S)", x_star=S_star, y_star=v_max,
                 title="Chemistry: Reaction velocity vs substrate (maximize over [0,50])")

cands_S = np.array([a, S_star, b])
cands_v = v_chem(cands_S)
j = int(np.argmax(cands_v))
print(f"Candidates (S): {cands_S}, v(S): {cands_v}")
print(f"Closed-interval maximum -> S = {cands_S[j]:.6f}, v = {cands_v[j]:.6f}")
```

## 3) Physics — Optics intensity vs distance (maximize intensity)

Model:  
$$
I(d) = \frac{500d}{(d+10)^2}, \quad 0 \le d \le 100.
$$

**Find $I'(d)$:**

Use the quotient rule:
$$
I'(d) = \frac{500(d+10)^2 - 500d \cdot 2(d+10)}{(d+10)^4}.
$$

Simplify:
$$
I'(d) = \frac{500(d+10)\big[(d+10) - 2d\big]}{(d+10)^4}
= \frac{500(d+10)(10-d)}{(d+10)^4}
= \frac{500(10-d)}{(d+10)^3}.
$$

Set $I'(d)=0$:
$$
500(10-d)=0 \;\;\Longrightarrow\;\; d=10.
$$

Thus, on the closed interval $[0,100]$, the **candidates** are
$$
d=0,\quad d=10,\quad d=100.
$$

**Evaluate $I$ at all candidates**

$$
I(0)=0, \qquad I(10)=\tfrac{500\cdot 10}{(10+10)^2}=\tfrac{5000}{400}=12.5, \qquad I(100)=\tfrac{500\cdot 100}{110^2}\approx 4.13.
$$

**Compare the values:**
$$
I(0)=0,\qquad I(10)=12.5,\qquad I(100)\approx 4.13.
$$

Therefore, the **absolute maximum** on $[0,100]$ occurs at
$$
\boxed{d^* = 10,\qquad I_{\max} = 12.5.}
$$

Candidates: $d\in\{0,10,100\}$. The interior critical point $d=10$ is the maximizer.

**Compare these results with the results from the python implemenation below:**

```python

def I_phys(d):
    return 500.0*d / (d + 10.0)**2

a, b = 0.0, 100.0
d_star, I_max, res_phys = report_max(I_phys, a, b, name="I")
plot_on_interval(I_phys, a, b, label="I(d)", x_star=d_star, y_star=I_max,
                 title="Physics: Intensity vs distance (maximize over [0,100])")

cands_d = np.array([a, d_star, b])
cands_I = I_phys(cands_d)
j = int(np.argmax(cands_I))
print(f"Candidates (d): {cands_d}, I(d): {cands_I}")
print(f"Closed-interval maximum -> d = {cands_d[j]:.6f}, I = {cands_I[j]:.6f}")
```

## 4) Statistics — Binomial likelihood (maximize likelihood)

Model (with $n=10$ trials, $k=7$ successes):  
$$
L(p) = \binom{10}{7} p^7 (1-p)^3, \quad 0 \le p \le 1.
$$

**Find $L'(p)$:**

Differentiate using the product/chain rule:
$$
L'(p) = \binom{10}{7}\Big[7p^6(1-p)^3 + p^7 \cdot 3(1-p)^2(-1)\Big].
$$

Simplify:
$$
L'(p) = \binom{10}{7}p^6(1-p)^2\big[7(1-p) - 3p\big]
= \binom{10}{7}p^6(1-p)^2(7 - 10p).
$$

Set $L'(p)=0$:
$$
p=0, \quad p=1, \quad \text{or} \quad 7-10p=0 \;\;\Longrightarrow\;\; p=0.7.
$$

Thus, on the closed interval $[0,1]$, the **candidates** are
$$
p=0,\quad p=0.7,\quad p=1.
$$

**Evaluate $L$ at all candidates**

$$
L(0)=0, \qquad L(0.7)=\binom{10}{7}(0.7)^7(0.3)^3 \approx 0.2668, \qquad L(1)=0.
$$

**Compare the values:**
$$
L(0)=0,\qquad L(0.7)\approx 0.2668,\qquad L(1)=0.
$$

Therefore, the **absolute maximum** on $[0,1]$ occurs at
$$
\boxed{p^* = 0.7,\qquad L_{\max} \approx 0.2668.}
$$

Candidates: $p\in\{0,0.7,1\}$. The interior critical point $p=0.7$ is the maximizer, which matches the intuitive estimator $p=k/n=0.7$.


**Compare these results with the results from the python implemenation below:**

```python

from math import comb

n, k = 10, 7
const = float(comb(n, k))

def L_stat(p):
    return const * (p**k) * ((1.0 - p)**(n - k))

a, b = 0.0, 1.0
p_star, L_max, res_stat = report_max(L_stat, a, b, name="L")
plot_on_interval(L_stat, a, b, label="L(p)", x_star=p_star, y_star=L_max,
                 title="Statistics: Binomial likelihood (maximize over [0,1])")

print(f"Analytic MLE p_hat = k/n = {k/n:.6f}")
cands_p = np.array([a, p_star, b])
cands_L = L_stat(cands_p)
j = int(np.argmax(cands_L))
print(f"Candidates (p): {cands_p}, L(p): {cands_L}")
print(f"Closed-interval maximum -> p = {cands_p[j]:.6f}, L = {cands_L[j]:.6f}")
```

## 5) Economics — Profit from demand and cost (maximize profit)

Demand $p(x)=100-2x$ on $0\le x\le 40$. Revenue $R(x)=100x-2x^2$.  
Cost $C(x)=20x+50$. Profit:  
$$
\pi(x) = 80x - 2x^2 - 50, \quad 0 \le x \le 40.
$$

**Find $\pi'(x)$:**

$$
\pi'(x) = 80 - 4x.
$$

Set $\pi'(x)=0$:
$$
80 - 4x = 0 \;\;\Longrightarrow\;\; x=20.
$$

Thus, on the closed interval $[0,40]$, the **candidates** are
$$
x=0,\quad x=20,\quad x=40.
$$

**Evaluate $\pi$ at all candidates**

$$
\pi(0) = -50, \qquad \pi(20) = 80(20) - 2(400) - 50 = 750, \qquad \pi(40) = 80(40) - 2(1600) - 50 = -270.
$$

**Compare the values:**
$$
\pi(0)=-50,\qquad \pi(20)=750,\qquad \pi(40)=-270.
$$

Therefore, the **absolute maximum** on $[0,40]$ occurs at
$$
\boxed{x^* = 20,\qquad \pi_{\max} = 750.}
$$

Candidates: $x\in\{0,20,40\}$. The interior critical point $x=20$ is the maximizer.


**Compare these results with the results from the python implemenation below:**

```python

def pi_econ(x):
    return 80.0*x - 2.0*x**2 - 50.0

a, b = 0.0, 40.0
x_star, pi_max, res_econ = report_max(pi_econ, a, b, name="pi")
plot_on_interval(pi_econ, a, b, label="pi(x)", x_star=x_star, y_star=pi_max,
                 title="Economics: Profit vs quantity (maximize over [0,40])")

cands_x = np.array([a, x_star, b])
cands_pi = pi_econ(cands_x)
j = int(np.argmax(cands_pi))
print(f"Candidates (x): {cands_x}, pi(x): {cands_pi}")
print(f"Closed-interval maximum -> x = {cands_x[j]:.6f}, pi = {cands_pi[j]:.6f}")
```
