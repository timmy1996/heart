```python
numerical_vars = list(Trip_Price_correlations.index)
numerical_data = data[numerical_vars]
```

```python
#attempt to find correlations among predictor variables
predictors_corr = correlations_data.drop('Trip_Price',axis=1).drop('Trip_Price',axis=0)
most_correlated = {}

for col in predictors_corr.columns:
    abs_corr = predictors_corr[col].drop(col).abs()
    
    # Find the variable with the highest absolute correlation
    most_corr_var = abs_corr.idxmax()
    most_corr_val = abs_corr.max()
    
    most_correlated[col] = (most_corr_var, most_corr_val)

# Convert to DataFrame for easier viewing
most_correlated_df = pd.DataFrame.from_dict(most_correlated, orient='index', columns=['Most Correlated With', 'Correlation (absolute val)'])
most_correlated_df = most_correlated_df.sort_values(by='Correlation (absolute val)', ascending=False)

print(most_correlated_df)
```

Above shows that predictors are not highly correlated(linearly, that is). During the modelling process we will not have to worry much about multicollinearity issues. The pair plot indicates as much.

```python
price_iqr = iqr(data['Trip_Price'])
third_quartile = data['Trip_Price'].describe()['75%']
ul = third_quartile  + 1.5 * price_iqr
potential_outliers = data[(data['Trip_Price'] > ul) ]
len(potential_outliers)
```

Just to check if the intuition is correct that the high, extreme prices correspond to longer distances. We create a scatter plot and identfy the outliers and it seems to be the case that the identified 'ouliers' corresponsd to the longer trip distances.

```python
categorical_vars = data.select_dtypes(include='object')
categorical_vars.head()
```

```python
cat_cols = categorical_vars.columns
for col in cat_cols:
    print(f"--- Summary for '{col}' ---")
    val_counts = data[col].value_counts(dropna=True)
    agg = data.groupby(col)['Trip_Price'].mean()
    summary = pd.concat([val_counts, agg], axis=1)
    summary.columns = ['Count', 'Mean Trip Price']
    print(summary)
    print("\n")
```

```python
n = len(cat_cols)
n_cols = 2

fig, axes = plt.subplots(n, n_cols, figsize=(12, 4 * n))
axes = axes.flatten()


for i, cat_var in enumerate(cat_cols):
    ax = axes[i]
    sns.kdeplot(data,x='Trip_Price',hue=cat_var, fill=True,common_norm=False,ax=ax)
    ax.set_title(f'Trip Price by {cat_var.replace('_', ' ')}')
    #ax.set_xlabel('Trip Price')
    ax.set_ylabel('Density')
    ax.grid(True)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(n, n_cols, figsize=(12, 4 * n))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    ax = axes[i]

    # Calculate mean trip price by category
    summary = data.groupby(col)['Trip_Price'].mean().sort_values(ascending=False)

    # Plot as a bar chart
    sns.barplot(x=summary.index, y=summary.values, ax=ax)

    ax.set_title(f'Mean Trip Price by {col.replace("_", " ")}')
    ax.set_xlabel(col.replace("_", " "))
    ax.set_ylabel('Mean Trip Price')
    ax.tick_params(axis='x', rotation=45)
    ax.grid(True)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(n, n_cols, figsize=(12, 4 * n))
axes = axes.flatten()
sns.set_style("white")  
for i, col in enumerate(cat_cols):
    ax = axes[i]

    # Calculate mean trip price by category
    summary = data.groupby(col)['Trip_Price'].mean().sort_values(ascending=True)  # ascending for nice horizontal layout

    # Horizontal bar plot: x = values, y = category labels
    sns.barplot(x=summary.values, y=summary.index, ax=ax)

    ax.set_title(f'Mean Trip Price by {col.replace("_", " ")}')
    ax.set_xlabel('Mean Trip Price')
    ax.set_ylabel(col.replace("_", " "))
    ax.grid(True)

# Hide unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

From the plots and table above, it appears the categorical variables have no major impact on the price, at least when looking at the mean sprice, suggesting they might not be that important wrt predicting the price. However, looking at traffic conditions, this variable seems to have a bit of impact, with high traffic indicating higher prices. This could be due to the fact that higher traffic could directly influence trip duration which is somewhat positively correlated with price (correlation of 0.22). 

```python
sns.scatterplot(data, x='Trip_Duration_Minutes', y='Trip_Price', hue='Traffic_Conditions')

plt.title('Scatter Plot of Trip Duration (mins) vs Trip Price by Traffic Conditions')
plt.xlabel('Trip Duration (mins)')
plt.ylabel('Trip Price')
```

Summary of findings 
- Trip price is highly correlated with trip distance (0.85), per kilometer rate (0.28) and trip duration (0.22) and to a lesser extent per minute rate. This suggests, among the continuous variables, these first two will prove important in the model building. 
- Categorical variables seem to not have a huge effect on the (mean) price, excpet for traffic condtions.
- There is a 'gap' in the scatter plot of distance vs price which could be a result of the missing values in the target variables. The effects on the performance of the model are yet to be seen.

```python
all_vars = list(data.columns)
preds = [var for var in all_vars if var != 'Trip_Price']
features = data[preds]
target = data['Trip_Price']
X, X_test, y, y_test = train_test_split(features, target,test_size=0.2, random_state=42,)
X.to_csv('../data/training_features.csv', index = False)
X_test.to_csv('../data/testing_features.csv', index = False)
y.to_csv('../data/training_labels.csv', index = False)
y_test.to_csv('../data/testing_labels.csv', index = False)
```
