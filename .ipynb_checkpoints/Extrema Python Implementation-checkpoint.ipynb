
# Maximizing Power on a Closed Interval Python Implementation

**Problem.** Consider question 7 in the MAT213 Problem set 1. The power delivered to a load is modeled by
$$
P(R) = \frac{100R}{(R+5)^2}, \qquad 0 \le R \le 200,
$$
where $R$ is the load resistance (in $\Omega)$.  
Use the **Closed Interval Method** to find the resistance that maximizes power and the maximum power delivered.



## Closed Interval Method (Recap)
On a closed interval $[a,b]$, a continuous function $f$ attains both a maximum and a minimum. To find the absolute extrema:
- Find **critical points** in $(a,b)$ (where $f'(x)=0$ or undefined).
- Evaluate $f$ at all **candidates**: the endpoints $a,b$ and any interior critical points.
- The largest value is the absolute maximum; the smallest is the absolute minimum.


## 1) Imports and model definition

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

def P(R):
    # Power delivered to the load as a function of resistance R (Ohms).
    return 100.0 * R / (R + 5.0)**2

# Interval
A, B = 0.0, 200.0
```

## 2) Quick behavior checks

```python

print(f"P(0)   = {P(0.0):.6f} W")
print(f"P(200) = {P(200.0):.6f} W")
# Note: P(R) is continuous for R â‰¥ 0; denominator never zero on [0,200].
```

## 3) Plot on the closed interval

```python

R = np.linspace(A, B, 100_001, endpoint=True)
Pvals = P(R)

i_max = int(np.argmax(Pvals))
R_star_num = float(R[i_max])
P_max_num  = float(Pvals[i_max])

plt.figure()
plt.plot(R, Pvals, label=r"$P(R)=\dfrac{100R}{(R+5)^2}$")
plt.axvline(5.0, linestyle="--", label=r"Analytic $R^* = 5\,\Omega$")
plt.scatter([R_star_num], [P_max_num], zorder=3, label="Grid max")
plt.xlim(A, B)
plt.xlabel(r"Load resistance $R$ (Î©)")
plt.ylabel(r"Power $P$ (W)")
plt.title("Power delivered vs. load resistance")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```


## 4) Analytical solution (calculus solution)
Compute the derivative:
$$
P'(R) \;=\; \frac{d}{dR}\!\left(\frac{100R}{(R+5)^2}\right)
= 100 \cdot \frac{(R+5)^2 - 2R(R+5)}{(R+5)^4}
= \frac{100\big(5 - R\big)}{(R+5)^3}.
$$
Critical point: $P'(R)=0 \iff R = 5$ (which lies in $(0,200)$.  
Second derivative test (or a sign check on $P'$) shows this is a maximum.  
Evaluate the candidates:
$$
P(0)=0, \quad P(200)=\frac{100\cdot 200}{205^2}\approx 0.4756,\quad P(5)=\frac{100\cdot 5}{10^2}=5.
$$
Therefore, the absolute maximum occurs at $R^*=5\,\Omega$ with $P_{\max}=5\,\text{W}$.


## 5) Numerical search (grid & SciPy)

This bit is for the extra curious. Here we use the function 
`scipy.optimize.minimize_scalar` to find the maximum of the power function.
Because `minimize_scalar` is a *minimizer*, we pass in `-P(r)` instead of `P(r)`. 
That way, minimizing `-P(r)` is the same as maximizing `P(r)`.

 As seen below, the `R_star_scipy` and `P_max_scipy` are both close to the calculus 
approach solution.


```python

# Grid-based maximization (already computed above as R_star_num, P_max_num)
print(f"Grid search -> R* â‰ˆ {R_star_num:.6f} Î©, Pmax â‰ˆ {P_max_num:.6f} W")

# SciPy bounded 1D maximization via minimizing the negative
res = minimize_scalar(lambda r: -P(r), bounds=(A, B), method="bounded")
R_star_scipy = float(res.x)
P_max_scipy  = float(P(R_star_scipy))

print(f"SciPy -> success={res.success}, evals={res.nfev}")
print(f"SciPy -> R* â‰ˆ {R_star_scipy:.6f} Î©, Pmax â‰ˆ {P_max_scipy:.6f} W")
```
